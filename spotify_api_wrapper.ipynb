```python
### Function to get tracks from playlist

def get_playlist_tracks(username, playlist_id):
    
    results = sp.user_playlist_tracks(username, playlist_id)
    tracks = results['items']
    
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    
    return tracks
```

```python
all_tracks = []
for p in playlists:
    tracks = get_playlist_tracks("spotify", p)
    all_tracks.extend(tracks)
```

```python
# Extract track name, track id, artist name, and artist id from a playlist

def extract4(tracks_dict):
    
    artist_id = []
    artist = []
    track = []
    track_id = []
    
    try:
        for i in range(0, len(tracks_dict)):
            artist_id.append(tracks_dict[i]["track"]["artists"][0]["id"])
            artist.append(tracks_dict[i]["track"]["artists"][0]["name"])
            track_id.append(tracks_dict[i]["track"]["id"])
            track.append(tracks_dict[i]["track"]["name"])
    except TypeError:
        pass
    
    dict = {"track_id": track_id, "track": track, "artist_id": artist_id, "artist": artist}
    return dict
```

```python
tracks_cleaned = extract4(all_tracks)
tracks_cleaned.keys()
```

```python
def audio_features(track_ids_list):
    audio_features = pd.DataFrame()
    for i, o in zip(range(0, (len(track_ids_list)+100), 100), range(100, (len(track_ids_list)+100), 100)):
        try:
            features = sp.audio_features(track_ids_list[i:o])
            features_df = pd.DataFrame(features)
            audio_features = pd.concat([audio_features, features_df])
        except AttributeError:
            pass
    return audio_features
```

```python
audio_features = audio_features(tracks_df["track_id"])
```
